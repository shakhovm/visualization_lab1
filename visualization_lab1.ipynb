```python
import pandas as pd
import altair as alt
```

```python
data_folder = "GLB.Ts_dSST.csv"
df = pd.read_csv(data_folder, skiprows=1)
df
```

```python
domain_pd = df["Year"].unique()[::10]
points = alt.Chart(df).mark_point(filled=True, 
                         color='#ff7b45', 
                         size=90)

left_years = df[df["Year"].isin([1904, 1998])]
right_years = df[df["Year"].isin([1944, 2014, 2015, 2017])]
max_year = df[df["J-D"] == df["J-D"].max()]
mean_ = df[(df["Year"] >= 1880) & (df["Year"] <= 1899)]["J-D"].mean()
right_text_year = alt.Chart(right_years).mark_text(
    align='left',
    baseline='middle',
    dx=7
).encode(
    text='Year'
)

left_text_year = alt.Chart(left_years).mark_text(
    align='right',
    baseline='middle',
    dx=-7
).encode(
    text='Year'
)

text_max_year = alt.Chart(max_year).mark_text(
    align='left',
    baseline='middle',
    dx=7,
    fontWeight='bold'
).encode(
    text='Year'
)

line = alt.Chart({"values": [{"x": 1880, "y": mean_}, {"x": 2017, "y": mean_}]}).mark_line(color='gray'
).encode(
    x='x:O',
    y ='y:Q' 
)

hot_text = alt.Chart({"values": [{"x": 2010, "y": mean_}]}).mark_text(
    text=['HOTTER THAN THE','1880-1899 AVERAGE'], fontSize=13, dy=-30, dx=-11
).encode(
    x='x:O', y='y:Q'
)

cold_text = alt.Chart({"values": [{"x": 2010, "y": mean_}]}).mark_text(
    text=['COLDER'], fontSize=13, dy=15, dx=20
).encode(
    x='x:O', y='y:Q'
)

al = (points + right_text_year + left_text_year + text_max_year + hot_text + cold_text).encode(
    x = alt.X(field='Year', 
              type = 'ordinal',
              title = 'Year',
              axis=alt.Axis(values=domain_pd,  labelAngle=0, grid=True, labelOffset=10, labelPadding=10)
             
              ),
    y = alt.Y(field = 'J-D', 
              type = 'quantitative',
              title='Temperature'
              )
)

(line + al).properties(
    width = 950, height = 400, background = '#F9F9F9', padding = 25,
    title=alt.TitleParams(
        ['Annual Global Surface Temperature,', 'Relative to Late 19th Century Avarage'],
        dx = -330,
        dy = 60,
        fontSize=15
    ))
```
